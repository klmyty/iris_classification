## 使用其他井数据进行测试

```python
import numpy as np
import pandas as pd
readpath = r'井8井3合并.csv'
data = pd.read_csv(readpath,encoding='GBK')
data.head()
```

```python
print("井3和8溢流标签\n",data['溢流'].value_counts())
```

## 使用onehotencoder

```python
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder()
encoded_kick= encoder.fit_transform(data[['溢流']])
# Create a DataFrame from the encoded data 
encoded_df = pd.DataFrame(encoded_kick.toarray(), columns=encoder.get_feature_names_out(['溢流']))
encoded_df = encoded_df.rename(columns={'溢流_0': 'kick_0', '溢流_1': 'kick_1', '溢流_2': 'kick_2'})

# Concatenate the original DataFrame and the encoded DataFrame
data = pd.concat([data, encoded_df], axis=1)
data.head()
```

```python
data.drop(columns=['溢流'], inplace=True)
```

### onehotencoder下划分X和y

```python
data_label=['kick_0','kick_1','kick_2']
```

```python
X = data.drop(data_label, axis=1)
y = data[data_label]
```

```python
X.head()
```

## 加载模型

```python
from joblib import load
# Load the saved model from disk
clf = load('random_forest_onehot.joblib')
```

```python
# Use the model for prediction
y_pred = clf.predict(X) 
```

```python
test_y = pd.DataFrame(y_pred)
```

```python
test_y.value_counts()
```

## 准确率，召回率、f1分数，看模型效果

```python
from sklearn.metrics import classification_report
print(classification_report(y, y_pred))
```
