# 合并井1井13
用清洗好的数据即可，将标签进行更改，可能溢流为0.5，油水侵为1，气侵为2，然后onehot编码即可！

```python
import numpy as np
import pandas as pd
```

```python
readpath = r'/home/klmyty/code/ML/data/溢流数据_clean/清洗后数据1_溢流.csv'
df = pd.read_csv(readpath,encoding='GBK')
df
```

```python
df['溢流'].value_counts()
```

```python
df['溢流'] = df['溢流'].astype(int)
df['溢流'].value_counts()
```

```python
readpath2 = r'/home/klmyty/code/ML/data/溢流数据_clean/清洗后数据13_溢流.csv'
df2 = pd.read_csv(readpath2,encoding='GBK')
df2.head()
```

## 查看两个井的大小

```python
print("井1数据大小",df.shape)
print("井13数据大小",df2.shape)
```

```python
print("井1溢流标签\n",df['溢流'].value_counts())
print("井13溢流标签\n",df2['溢流'].value_counts())
```

## 井1为水侵，井13为气侵 故井1的1还为1，井3的1为2

```python
df2.loc[df2['溢流']==1,'溢流'] = 2
```

```python
print("井13溢流标签\n",df2['溢流'].value_counts())
```

# 合并井1井13

```python
data = pd.concat([df, df2], join="inner")

# Select the columns you want to keep
data = data[['井深(8004) m', '迟到井深(8006) m', '大钩高度(8008) m', '全烃(8219) %', '入口温度(8106) DEG.C', '泵冲次1(8059) spm', '出口密度(8105) g/cm3', '套管压力(8019) MPa', '钻时(8020) min/m', '立管压力(8018) MPa', '钻压(8011) kN', '入口流量(8110) L/s', '出口温度(8107) DEG.C', '钻头位置(8005) m', '入口密度(8104) g/cm3', '扭矩(8013) kN.m', '大钩负荷(8010) kN', '总池体积(8112) m3', '溢流', '转盘转速(8012) RPM']]
print(data)
```

判断是否有列的数值一直没有变化

```python
for column in data.columns:
    if data[column].value_counts().count() == 1:
        print(column)
```

发现没有一直没有变化的列

变化较少的列，少于3的列

```python
for column in data.columns:
    number = data[column].value_counts().count()
    if number <= 3:
        print(column,number)
```

可以看到也没有变化较少的列

保存合并好的井1井13数据

```python
data.to_csv('井1井13合并.csv', index=False, mode='w', header=True,encoding='GBK')
```
