# 数据编码onehot

```python
import numpy as np
import pandas as pd
readpath = r'井1井3合并.csv'
data = pd.read_csv(readpath,encoding='GBK')
data.head()
```

## 标签编码 对溢流和工况都进行编码

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
data['溢流'] = le.fit_transform(data['溢流'])
# data['工况'] = le.fit_transform(data['工况'])
data.head()
```

```python
data['溢流'].value_counts()
```

## 保存被编码后的数据

保存了无工况被编码的 井1井3合并_encoded

有工况被编码的 井1井3合并_encodedwithconditon.csv

```python
data.to_csv('井1井3合并_encodedwithLabelEncoder.csv', index=False, mode='w', header=True,encoding='GBK')
```
