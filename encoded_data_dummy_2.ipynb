## 找到object类型的特征

```python
# select the columns that are of type object
object_columns = data.select_dtypes(include=['object'])

# get the column names
object_column_names = object_columns.columns.tolist()

print("Columns of type object:", object_column_names)
```

## 标签编码 onehot 对溢流和工况都进行编码

data 数据

prefix='kick' 列的前缀

columns=['溢流'] 要被编码的列

## get_feature_names_out 方法用于根据原始列名称检索单热编码列的名称。

```python
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder()
encoded_kick= encoder.fit_transform(data[['溢流']])
# Create a DataFrame from the encoded data 
encoded_df = pd.DataFrame(encoded_kick.toarray(), columns=encoder.get_feature_names_out(['溢流']))
encoded_df = encoded_df.rename(columns={'溢流_0': 'kick_0', '溢流_1': 'kick_1', '溢流_2': 'kick_2'})

# Concatenate the original DataFrame and the encoded DataFrame
df = pd.concat([data, encoded_df], axis=1)
df.head()
```

```python
data.drop(columns=['时间','溢流'], inplace=True)
```

```python
from joblib import dump
# Save the model to disk
dump(encoded_kick, 'onehotencoder.joblib')
```
