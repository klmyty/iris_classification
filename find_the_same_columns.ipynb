```python
import glob


# Use glob to find all CSV files in the folder
path = '/home/klmyty/code/ML/data/溢流数据_clean/*.csv'
csv_files = glob.glob(path)

# Read each CSV file into a separate dataframe
dfs = [pd.read_csv(file,encoding='GBK') for file in csv_files]
```

intersection 它返回两个或多个集合之间的公共元素。公共元素在新集合中返回。

```python
# Find the common columns between dataframes in the list
common_columns = set(dfs[0].columns).intersection(*(set(df.columns) for df in dfs))

print(f"The common columns between the dataframes are: {common_columns}")
```
